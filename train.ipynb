```python
def categorical_numeric_features(train_ds):

    categorical_features = []
    numeric_features = []
    
    for column_name in train_ds.columns:
        if train_ds[column_name].dtype == 'object':
            categorical_features.append(column_name)
        else:
            numeric_features.append(column_name)
    return categorical_features , numeric_features 
```

```python
def encode_categorical_features(train_ds, categorical_features):

  encoder = OneHotEncoder(sparse_output=False, drop='first')
  encoded_data = encoder.fit_transform(train_ds[categorical_features])
  encoded_feature_names = encoder.get_feature_names_out(categorical_features)
  encoded_df = pd.DataFrame(encoded_data, columns=encoded_feature_names)
  df_encoded = pd.concat([train_ds.drop(columns=categorical_features), encoded_df], axis=1)

  return df_encoded
```

```python
categorical_features , numeric_features = categorical_numeric_features(train_ds)
df_encoded = encode_categorical_features(train_ds ,categorical_features )
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
```

```python
def results_models(x_train, y_train, x_test, y_test):
    
    
    results = []
    
    # Linear Regression
    model = LinearRegression()
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    results.append({
          'model': 'linearregression',
          'mse': mean_squared_error(y_test, predictions),
          'mae': mean_absolute_error(y_test, predictions),
          'r2': r2_score(y_test, predictions)
      })
    
    # Linear Regression with Polynomial Features
    poly_transformer = PolynomialFeatures(degree=2)
    X_train_poly = poly_transformer.fit_transform(x_train)
    X_test_poly = poly_transformer.transform(x_test)
    model = LinearRegression()
    model.fit(X_train_poly, y_train)
    predictions = model.predict(X_test_poly)
    results.append({
          'model': 'linearregression_poly_f',
          'mse': mean_squared_error(y_test, predictions),
          'mae': mean_absolute_error(y_test, predictions),
          'r2': r2_score(y_test, predictions)
      })
    
    # Decision Tree Regressor
    model = DecisionTreeRegressor()
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    results.append({
          'model': 'DecisionTreeRegressor',
          'mse': mean_squared_error(y_test, predictions),
          'mae': mean_absolute_error(y_test, predictions),
          'r2': r2_score(y_test, predictions)
      })
    
    # Random Forest Regressor
    model = RandomForestRegressor(n_estimators=100)
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    results.append({
          'model': 'RandomForestRegressor',
          'mse': mean_squared_error(y_test, predictions),
          'mae': mean_absolute_error(y_test, predictions),
          'r2': r2_score(y_test, predictions)
      })
    
    # SVR
    model = SVR(kernel='rbf', C=1.0)
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    results.append({
          'model': 'SVR',
          'mse': mean_squared_error(y_test, predictions),
          'mae': mean_absolute_error(y_test, predictions),
          'r2': r2_score(y_test, predictions)
      })
    
    # KNeighborsRegressor
    model = KNeighborsRegressor(n_neighbors=21)
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    results.append({
          'model': 'KNeighborsRegressor',
          'mse': mean_squared_error(y_test, predictions),
          'mae': mean_absolute_error(y_test, predictions),
          'r2': r2_score(y_test, predictions)
      })
    
    # Gradient Boosting Regressor
    model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1)
    model.fit(x_train , y_train)
    predictions = model.predict(x_test)
    results.append({
          'model': 'GradientBoostingRegressor',
          'mse': mean_squared_error(y_test, predictions),
          'mae': mean_absolute_error(y_test, predictions),
          'r2': r2_score(y_test, predictions)
      })
    return results 
```

```python
reults = results_models(x_train, y_train, x_test, y_test)
pd.DataFrame(reults)
```
